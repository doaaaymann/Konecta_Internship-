## Load and Preview the Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("supermarket_sales.csv")
```

## Clean Column Names (Removing spaces and Converting to lowercase)

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
df.head()
```

## Convert Date 

```python
print(df["date"].unique())
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

## Convert Time Columns

```python
df.time
```

```python
df["time"] = pd.to_datetime(df["time"], format='%H:%M').dt.strftime("%H:%M")
```

```python
df.head()
```

##  Create Time-Based Features (Day, Month, Week Number)

```python
df["date"] = pd.to_datetime(df["date"], errors='coerce')
```

```python
df["Day_of_week"] = df["date"].dt.day_name()
```

```python
df["Month_Name"] = df["date"].dt.month_name()
```

```python
df["Week_number"] = df["date"].dt.isocalendar().week
```

```python
df.head()
```

## Handle Missing Values

```python
df.info()
```

```python
print((df.isnull().sum() / len(df)) * 100)
```

## Analyze Sales Trends Over Time



```python
daily_sales = df.groupby('date')['total'].sum()
print(daily_sales.head())
```

```python
monthly_sales = df.groupby(df['date'].dt.to_period('M'))['total'].sum()
print(monthly_sales.head())
```

## Analyze Sales by Hour

```python
df['hour'] = pd.to_datetime(df['time'], format='%H:%M').dt.hour
```

```python
sales_by_hour = df.groupby('hour')['total'].sum()
```

```python
sales_by_hour.head()
```

## Compare Spending by Gender

```python
spending_by_gender = df.groupby('gender')['total'].agg(['sum', 'mean', 'count'])
```

```python
df.head()
```

##  Analyze Product Ratings

```python
avg_rating = df.groupby('product_line')['rating'].mean()
```

## Analyze Payment Method Distribution
